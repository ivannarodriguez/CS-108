## Background

I used RStudio to do some data wrangling to select the my features of interest according to what the owner of the dataset thought were important predictors of coffee productivity. I saved the resulting dataset after the data transformations in a new file called `coffee2.csv`. 

I decided to do exploratory data analysis to see the kinds of relationships present between my features and label, and to see whether there are many outliers present in some of the features. My label is `productividad` which is the variable that reports the coffee productivity. I will use the `seaborn` library to plot my features against my predictor in the following chunks of code below.

```
# load libraries
import pandas as pd
import seaborn as sns
```

```
# Loading in the data and changing numeric to categorical
url = 'https://raw.githubusercontent.com/ivannarodriguez/cs344/master/project/coffee2.csv'
df = pd.read_csv(url, encoding = "UTF-8")
col_names = ['Tieneaccesoalcrédito', 'Hacefertilizacionquimica', 'Hacefertilizacionorganica', 'Municipio', 'Sexo']
for col in col_names:
  df[col] = df[col].astype('category',copy=False)

df.dtypes
```

```
# EDA
import numpy as np
m = sns.boxplot(y='Municipio', x='productividad', data=df, whis=np.inf, color='white')
m = sns.stripplot(y='Municipio', x='productividad', data=df, color=".3", alpha = 0.5).set_title('Productivity by Municipality')
```

The plot above shows the relationship between the municipality `Municipio` in which the coffee is grown and the productivity, `productividad`, measured in number of quintals per manzana (0.7 acres). From this boxplot I can see that there are many outliers, especiallly in Danli. There also seems to be differences among municipalities, and more data recorded for some than for others.

```

s = sns.boxplot(x='Sexo', y='productividad', data=df, whis=np.inf, color='white')
s = sns.stripplot(x='Sexo', y='productividad', data=df, color=".3", alpha = 0.2).set_title('Productivity by Gender')
```

There does not seem to be a difference in the amount of coffee produced between men and women according to the data, since the medians are about the same. It is slightly lower for women, however, and there are less datapoints for women too. There is one very apparent outlier for the men.

```
c = sns.boxplot(x='Tieneaccesoalcrédito', y='productividad', data=df, whis=np.inf, color='white')
c = sns.stripplot(x='Tieneaccesoalcrédito', y='productividad', data=df, color=".3", alpha = 0.2).set_title('Productivity by Access to Credit')
```

Though both medians seem to be the same for the boxplots above, there seems to be a lot more people within the 20-40 productivity range for those who have access to credit. There could be a relationship between productivity and credit although it is not very apparent.
